```
from matplotlib import pyplot as plt
from sklearn import datasets
import numpy as np
```

```
n_samples = 60
X, y = datasets.make_regression(n_samples = n_samples, n_features = 0, noise = 0.005, random_state = 2)
X = range(60)
plt.scatter(X, y)
plt.show()
```

```
def plot_ewa(beta):
    v, bias_corrected, X_avg, X_avg_bias_corrected = 0, 0, [], []
    for i in range(n_samples):
        v = beta * v + (1.0 - beta) * y[i]
        v_bias_corrected = v / (1.0 - np.power(beta, i+1))
        X_avg.append(v)
        X_avg_bias_corrected.append(v_bias_corrected)
    plt.scatter(X, y)
    plt.plot(X, X_avg, '.r-')
    plt.plot(X, X_avg_bias_corrected, '.g-') # bias corrected
    plt.show()
```

```
for beta in range(0, 5):
    plot_ewa(beta * 0.1)
```
