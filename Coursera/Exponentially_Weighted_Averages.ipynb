```
from matplotlib import pyplot as plt
from sklearn import datasets
```

```
n_samples = 60
X, y = datasets.make_regression(n_samples = n_samples, n_features = 0, noise = 0.005, random_state = 1)
X = range(60)
plt.scatter(X, y)
plt.show()
```

```
def plot_ewa(beta):
    v, X_avg = 0, []
    for i in range(n_samples):
        v = beta * v + (1 - beta) * y[i]
        X_avg.append(v)
    plt.scatter(X, y)
    plt.plot(X, X_avg, '.r-')
    plt.show()
```

```
for beta in range(11):
    plot_ewa(beta * .1)
```
